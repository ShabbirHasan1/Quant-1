```python
%matplotlib inline
```

```python
df = pd.read_csv('walmart_stock.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.set_index('Date', inplace= True)
```

```python
df.head()
```

```python
df.resample(rule='BQ').mean()
```

```python
df['Open'].plot(figsize = (16,8))
```

```python
df.rolling(window=7).mean().head(20)
# Moving Average
```

```python
df['Open'].plot()
df['Open'].rolling(30).mean().plot(figsize=(16,8))
```

```python
df['Close 30 days MA'] = df['Close'].rolling(30).mean()
df[['Close 30 days MA', 'Close']].plot(figsize=(16,8), grid = 1, lw=3)
```

```python
df['Close'].expanding().mean().plot(figsize=(16,8),grid=1,lw=4)
```

```python
# Bollinger Bands

# Close 20 Day MA

df['Close 20 Day MA'] = df['Close'].rolling(window=20).mean()

# Upper = 20 MA + 2*(std over 20 days)

df['Upper'] = df['Close 20 Day MA'] + 2*(df['Close'].rolling(20).std())

# Lower = 20 MA - 2*(std over 20 days)

df['Lower'] = df['Close 20 Day MA'] - 2*(df['Close'].rolling(20).std())
```

```python
df[['Close','Close 20 Day MA','Upper','Lower']].tail(200).plot(figsize=(18,6),grid=0)
```
