```python
import pandas as pd, numpy as np
```

```python
mcdon = pd.read_csv('mcdonalds.csv', index_col='Date', parse_dates=True)
```

```python
mcdon.head()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(15,4))
```

```python
mcdon['Adj. Volume'].plot(xlim = ['2007-01-01', '2009-01-01'],ylim=[0,0.6*(10**8)], figsize = (17,4), grid = 1)
```

```python
import matplotlib.dates as dates
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

# LOCATING
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATTING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
```
